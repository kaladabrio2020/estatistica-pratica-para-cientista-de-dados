# Distribuiçaõ de dados e amostras

**É uma equivoco comum é pensar que a era do big data significa o fim da necessidade de amostragem**. Na verdade, com a proliferação de dados de qualidade e relevancias variaveis a necessidade da amostragem como ferramente para trabalhar bem com uma variedade de dados e minimizar viés.
> Mesmo com o grande quantidade de dados, **treinar um modelo preditivo custuman ser desenvolidos e conduzidos com amostra**

## Amostragem aleatoria e vies de amostra

* **Amostra:** um subconjunto de um conjunto maior de dados.
* $N$ : Tamanho da população
* **Viés** : Erro sistematico
* **Viés de amostragem** : Uma amostra não representa a população

> O termo não aleatorio é importante = dificilmente uma amostra, inclusive aleatoria, será exatamente representativa da população. O **vies de amostragem ocorre quando a diferença é significativa e pode se propagar em outras amostras extraídas da mesma forma que a primeira**

### Viés de selecao

Se refere a pratica de escolher os dados de forma seletiva - consciente ou não - de modo que isso leva a uma conclusao que é enganosa ou efemera.
